```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("world-country-populations.csv")
```

```python
#checking for missing values
df.isnull().sum()
```

```python
#getting the statistical measures of the dataset
df.describe()
```

```python
# Convert necessary columns to string type
df['city'] = df['city'].astype(str)
df['country'] = df['country'].astype(str)
df['type'] = df['type'].astype(str)
```

```python
#countries with highest population in 2022

# Sort the DataFrame by population_in_2022 column in descending order
sorted_df = df.sort_values(by='population_in_2022', ascending=False)


# Get the top 10 countries
top_10_countries = sorted_df.head(10)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(top_10_countries['country'], top_10_countries['population_in_2022'])
plt.xlabel('Country')
plt.ylabel('Population in 2022')
plt.title('Top 10 Countries with Highest Population in 2022')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Top 10 Countries with Highest Population in 2022")
plt.show()
```

```python
#countries with highest population in 2023

# Sort the DataFrame by population_in_2023 column in descending order
sorted_df = df.sort_values(by='population_in_2023', ascending=False)

# Get the top 10 countries
top_10_countries = sorted_df.head(10)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(top_10_countries['country'], top_10_countries['population_in_2023'], color = 'coral')
plt.xlabel('Country')
plt.ylabel('Population in 2023')
plt.title('Top 10 Countries with Highest Population in 2023')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig("Top 10 Countries with Highest Population in 2023")
```

```python
# Sort the DataFrame by population_in_2022 column in descending order
sorted_df = df.sort_values(by='population_in_2022', ascending=False)

# Get the top 10 cities
top_10_cities = sorted_df.head(10)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(top_10_cities['city'], top_10_cities['population_in_2022'], color = 'burlywood')
plt.xlabel('City')
plt.ylabel('Population in 2022')
plt.title('Top 10 Cities with Highest Population in 2022')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig("Top 10 Cities with Highest Population in 2022")
```

```python
# Sort the DataFrame by population_in_2023 column in descending order
sorted_df = df.sort_values(by='population_in_2023', ascending=False)

# Get the top 10 cities
top_10_cities = sorted_df.head(10)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(top_10_cities['city'], top_10_cities['population_in_2023'], color = 'darkolivegreen')
plt.xlabel('City')
plt.ylabel('Population in 2023')
plt.title('Top 10 Cities with Highest Population in 2023')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.savefig('Top 10 Cities with Highest Population in 2023')
```

```python
#population growth rate of countries

# Sort the DataFrame by population_growthRate column in descending order
sorted_df = df.sort_values(by='population_growthRate', ascending=False)

# Plotting
plt.figure(figsize=(12, 8))
plt.bar(sorted_df['country'], sorted_df['population_growthRate'], color='indigo')
plt.xlabel('Country')
plt.ylabel('Population Growth Rate')
plt.title('Population Growth Rate by Country')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
plt.savefig('Population Growth Rate by Country')
```

```python
#top 20 cities by growth rate

# Sort the DataFrame by population_growthRate column in descending order
sorted_df = df.sort_values(by='population_growthRate', ascending=False)

# Get the top 20 cities
top_20_cities = sorted_df.head(20)

# Plotting
plt.figure(figsize=(12, 8))
plt.bar(top_20_cities['city'], top_20_cities['population_growthRate'], color='skyblue')
plt.xlabel('City')
plt.ylabel('Population Growth Rate')
plt.title('Top 20 Cities by Population Growth Rate')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
plt.savefig("'Top 20 Cities by Population Growth Rate'")
```

```python
# bottom  20 cities by population growth rate

# Sort the DataFrame by population_growthRate column in descending order
sorted_df = df.sort_values(by='population_growthRate', ascending=True)

# Get the bottom 20 cities
bottom_20_cities = sorted_df.head(20)

# Plotting
plt.figure(figsize=(12, 8))
plt.bar(bottom_20_cities['city'], bottom_20_cities['population_growthRate'], color='skyblue')
plt.xlabel('City')
plt.ylabel('Population Growth Rate')
plt.title('Bottom 20 Cities by Population Growth Rate')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
plt.savefig('Bottom 20 Cities by Population Growth Rate')
```

```python

# Group by country and sum up population for 2022 and 2023
grouped_df = df.groupby('country')[['population_in_2022', 'population_in_2023']].sum()

# Plotting
fig, ax = plt.subplots(figsize=(12, 8))
grouped_df.plot(kind='bar', stacked=True, ax=ax)
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population Comparison between 2022 and 2023 by Country')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
plt.savefig('Population Comparison between 2022 and 2023 by Country')
```

```python
#interactive version for the above plot


# Group by country and sum up population for 2022 and 2023
grouped_df = df.groupby('country')[['population_in_2022', 'population_in_2023']].sum().reset_index()

# Melt the dataframe to long format for Plotly
melted_df = pd.melt(grouped_df, id_vars='country', var_name='year', value_name='population')

# Plotting
fig = px.bar(melted_df, x='country', y='population', color='year', barmode='stack',
             labels={'population': 'Population', 'country': 'Country', 'year': 'Year'},
             title='Population Comparison between 2022 and 2023 by Country')
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['population_in_2022'], df['population_growthRate'], color='blue', alpha=0.5)
plt.xlabel('Population in 2022')
plt.ylabel('Population Growth Rate')
plt.title('Population Growth Rate vs Population in 2022')
plt.grid(True)
plt.tight_layout()
plt.show()
plt.savefig('Population Growth Rate vs Population in 2022')
```

```python
# Filter the DataFrame to include only cities where type is 'us'
us_cities_df = df[df['type'] == 'us']

# Group by city and sum up population for 2022 and 2023
grouped_df = us_cities_df.groupby('city')[['population_in_2022', 'population_in_2023']].sum()

# Plotting stacked bar plot for population comparison
plt.figure(figsize=(12, 8))
grouped_df.plot(kind='bar', stacked=True)
plt.xlabel('City')
plt.ylabel('Population')
plt.title('Population Comparison between 2022 and 2023 for Cities in the USA')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig("Population Comparison between 2022 and 2023 for Cities in the USA")
```

```python
# Plotting growth rate for cities in the USA
plt.figure(figsize=(10, 6))
plt.bar(us_cities_df['city'], us_cities_df['population_growthRate'], color='green')
plt.xlabel('City')
plt.ylabel('Population Growth Rate')
plt.title('Population Growth Rate for Cities in the USA')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig("Population Growth Rate for Cities in the USA")
```
